```python
import pandas as pd
```

```python
form_users = {
    'User ID': [1, 2, 3, 4, 5],
    'Username': ['bogdan_s', 'Aidyn', 'Saltanat', 'Zhangir', 'bob88'],
    'Age': [18, 24, 20, 14, None],
    'Joined Date': pd.to_datetime(['2020-12-15', '2022-01-05', '2034-02-20', '2000-06-15', '2023-11-29']),
    'Total Post': [150, 200, 354, 500, 290],
    'Reputation': [500, 2000, 1243, 453, 680]
}

df = pd.DataFrame(form_users)
```

# Filtration

```python
df[df['Age'] >= 20]
```

```python
df[df['Total Post'] >= 300]
```

```python
df[df['Reputation'] == 500]
```

```python
df[(df['Total Post'] >= 300) & (df['Age'] >= 20)]
```

```python
df[(df['Total Post'] >=400) | (df['Age'] == 24)]
```

```python
df['Reputation'].isin([200, 1243])
```

```python
df[df['Reputation'].isin([500, 1243])]
```

```python
df[df['Reputation'].isin(range(200, 501))]
```

# Слияние датафреймов

```python
total_comments = {
    'Total comments': [70, 30, 45, 80, 24]
}
df2 = pd.DataFrame(total_comments)
df2
```

```python
# По умолчанию метод concat соединяет по оси 0!
pd.concat([df, df2])
```

```python
df = pd.concat([df, df2], axis=1)
df
```
